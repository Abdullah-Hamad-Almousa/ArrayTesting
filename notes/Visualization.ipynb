```python
from os.path import split

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

import warnings

warnings.filterwarnings("ignore")
```

## Using Matplotlib and Seaborn it Will be fun

```python
df = pd.read_csv("coffeeSales.csv")
df.head(3)
```

Change the names of the columns to camelCapital

```python
df = df.rename(columns={"Customer_ID": "ID", "Unit Price": "UnitPrice", "Sales Amount": "SalesAmount",
                         "Used_Discount": "UsedDiscount", "Discount_Amount": "DiscountAmount", "Final Sales": "FinalSales"})
df.head(3)
```

```python
df.info()
```

```python
df.nunique()
```

```python
df.drop('Category', axis=1, inplace=True)
df.head(3)
```

```python
df.Product.unique()
```

```python
df['UsedDiscount'] = df['UsedDiscount'].astype(int)
df.head(3)
```

```python
df['Date'] = pd.to_datetime(df['Date'])
df.info()
```

```python
df.head(3)
```

```python
df.City.unique()
```

```python
x = np.linspace(1, 20, 100)
y = np.exp(x)

plt.plot(x, y)

plt.show()
```

```python
plt.plot(x, y, 'b^') # b for blue can change to green or red ||  ^ for the dots can change it to *

plt.show()
```

```python
y1 = x
y2 = np.square(x)
y3 = np.sqrt(x)

plt.plot(x, y1, 'y--', x, y2, 'b^', x, y3, 'g*')
plt.title('Line Plot')
plt.show()
```

# Scatter

```python
x = df['SalesAmount']
y = df['FinalSales']

plt.scatter(x,y, color='red')

plt.xlabel('City')
plt.ylabel('Sales')

plt.show()
```

```python
plt.scatter(x,y, s = np.array(y)*.2, c="green", alpha=0.8)

plt.title('Scatter Plot')

plt.xlabel('City')
plt.ylabel('Sales')

plt.show()
```

# Pie

```python
plt.pie(df.Product.value_counts(), autopct='%.1f%%', radius=1.2)
plt.title('Pie Plot')
plt.show()
```

```python
plt.pie(df.Product.value_counts(), autopct='%.1f%%', radius=1.2, explode=(0, 0, 0, 0, 0.5))
plt.title('Pie Plot')
plt.show()
```

```python
plt.pie(df.Product.value_counts(), autopct='%.1f%%', radius=1.2)

circle = plt.Circle((0, 0), .5, color='white')
plot = plt.gcf()
plot.gca().add_artist(circle)

plt.title('Pie Plot')
plt.show()
```

# Histogram

```python
plt.hist(df['FinalSales'], bins=5)

plt.title('Histogram Plot')
plt.xlabel('Final Sales')
plt.ylabel('Frequency')

plt.show()
```

# Count plot

```python
sns.countplot(data = df, x = 'City', hue = 'FinalSales')

plt.show()
```

# Box plot

```python
plt.boxplot(df['FinalSales'])

plt.text(x = 1.1, y = df['FinalSales'].min(), s = 'min')
plt.text(x = 1.1, y = df.FinalSales.quantile(.25), s = 'Q1')
plt.text(x = 1.1, y = df['FinalSales'].median(), s = 'Q2')
plt.text(x = 1.1, y = df.FinalSales.quantile(.75), s = 'Q3')
plt.text(x = 1.1, y = df['FinalSales'].max(), s = 'max')

plt.title('Box Plot of final sales')
plt.ylabel('Total Sales')

plt.show()
```

```python
plt.boxplot(df['FinalSales'])

plt.text(x = 1.1, y = df['FinalSales'].min(), s = 'min')
plt.text(x = 1.1, y = df.FinalSales.quantile(.25), s = 'Q1')
plt.text(x = 1.1, y = df['FinalSales'].median(), s = 'Q2')
plt.text(x = 1.1, y = df.FinalSales.quantile(.75), s = 'Q3')
plt.text(x = 1.1, y = df['FinalSales'].max(), s = 'max')
#                        Where to point to| Where to the text to be display
plt.annotate('Outliers', xy = (.97, 2005), xytext = (.7, 1500), arrowprops = dict(arrowstyle = 'simple', facecolor = 'red')) # -> There is no outliers at all just exploring

plt.title('Box Plot of final sales')
plt.ylabel('Total Sales')

plt.show()
```
