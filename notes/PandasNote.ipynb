```python
df = pd.read_csv("bmi.csv")
df.head(3)
```

```python
df = df.rename(columns={"Gender": "Male"})

genderMap = {'Male': 1, 'Female': 0}
df["Male"] = df["Male"].map(genderMap)
df.head(3)
```
