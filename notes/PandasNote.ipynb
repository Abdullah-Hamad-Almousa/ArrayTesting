```python
Subject = pd.Series(['Maths', 'Science', 'Python', 'English', 'Religion', 'Java'])
Marks = pd.Series([78, 89, 79, 65, 77, 99])
CGPA = pd.Series([3.5, 4, 5, 2.5, 3.5, 5])
```

```python
GatherData = pd.DataFrame([Subject, Marks, CGPA], index=['Subject', 'Marks', 'CGPA'])
GatherData
```

```python
GatherData = pd.DataFrame([Subject, Marks, CGPA], index=['Subject', 'Marks', 'CGPA']).T
GatherData
```
