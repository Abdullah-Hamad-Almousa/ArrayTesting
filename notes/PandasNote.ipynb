```python
elements = [-1, 2, -3, -4, 5]

def custom_function(x):
    return x if x > 0 else 0

series = pd.Series(elements)

transformed_series = series.apply(custom_function)

zero_count = (transformed_series == 0).sum()

print(series, "\n\n", transformed_series, "\n\n", zero_count)
```
