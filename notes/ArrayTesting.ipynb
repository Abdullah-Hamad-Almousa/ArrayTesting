```python
a = np.array([0, np.pi/6, np.pi/4, np.pi/3, np.pi/2])

sin_a = np.sin(a)

cos_complement_a = np.cos(np.pi/2 - a)

# Validate the identity (sin(x) ≈ cos(π/2 - x))
complementary_identity_check = np.allclose(sin_a, cos_complement_a)

print(a, "\n",sin_a, "\n", cos_complement_a, "\n", complementary_identity_check)
```

```python
a = np.array([np.pi/6, np.pi/4, np.pi/3, np.pi/2, np.pi])

cos_double_a = np.cos(2*a)

cosine_double_angle_identity = np.cos(a)**2 - np.sin(a)**2

identity_check = np.allclose(cosine_double_angle_identity, cos_double_a)

identity_check
```

```python
a = np.array([np.pi/6, np.pi/4, np.pi/3, 5*np.pi/6, 7*np.pi/6])

tan_a = np.tan(a)

cot_a = 1/tan_a

expected_cot_a = np.cos(a) / np.sin(a)
reciprocal_check = np.allclose(cot_a, expected_cot_a)

print(a, "\n", tan_a, "\n", cot_a, "\n", expected_cot_a, "\n", reciprocal_check)
```
